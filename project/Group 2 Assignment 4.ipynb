```python
class node:
    def __init__(self,data= None):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = node()
    
    def addFirst(self, data):
        newdata = node(data)
        newdata.next = self.head.next
        self.head.next = newdata
        
        
    def addLast(self, data):
        cur = self.head
        while cur.next != None:
            cur = cur.next
        cur.next = node(data)
        
    def getFirst(self):
        if self.head.next == None:
            return
        
        return self.head.next.data
    
    def getLast(self):
        if self.head.next == None:
            return
        
        cur = self.head
        while cur.next != None:
            cur = cur.next
        
        return cur.data
    
    def removeFirst(self):
        self.head = self.head.next
    
    def removeLast(self):
        if self.head.next == None:
            return
        
        cur = self.head
        prev = None
        
        while cur.next != None:
            prev = cur
            cur = cur.next
            
        prev.next = None
        return prev
            
    
    def add(self, data):
        self.addLast(data)
        
    def insert(self, value, index):
        if self.head.next == None:
            return 'List is empty'

        cur_index = 0
        cur = self.head
        while cur != None:
            if cur_index == index:
                new_node = node(value)
                new_node.next = cur.next
                cur.next = new_node
                return
            cur = cur.next
            cur_index += 1
        return 'Out Of Range'
    
    def clear(self):
        self.head.next = None
        return self.head
    
    def contains(self,value):
        if self.head.next == None:
            return False
        
        cur = self.head.next
        
        while cur != None:
            if  cur.data == value:
                return True
            cur = cur.next
        
        return False

    
    def get(self,index):
        if self.head.next == None:
            return False
        
        cur = self.head.next
        cur_index = 0

        while cur != None:
            if cur_index == index:
                return cur.data
            cur = cur.next
            cur_index += 1

        return "Out Of Range"
        
    
    def indexOf(self,value):
        if self.head.next == None:
            return 'List is empty'
        
        cur = self.head.next
        cur_index = 0
        
        while cur != None:
            if value == cur.data:
                return cur_index
            cur = cur.next
            cur_index += 1
        return 'Value is not present'
    
    def isEmpty(self):
        return False if self.head.next != None else True
    
    def lastIndexOf(self,value):
        if self.head.next == None:
            return 'List is empty'
    
        cur = self.head.next
        last_index = -1 # initialize to -1 since we haven't found the value yet
        cur_index = 0

        while cur is not None:
            if value == cur.data:
                last_index = cur_index
            cur = cur.next
            cur_index += 1

        if last_index == -1:
            return 'Value is not present'
        else:
            return last_index
    
    def getSize(self):
        cur  = self.head
        count = 0
        while cur.next != None:
            count += 1
            cur = cur.next
        return count
    
    def remove(self, value):
        if self.head.next == None:
            return 'List is empty'
        cur = self.head
        prev = None
        
        while cur.next != None:
            if value == cur.data:
                if prev == None:
                    self.head = cur.next
                else:
                    prev.next = cur.next
                return 
            prev = cur
            cur = cur.next
    
    def removeAt(self,index):
        if self.head.next == None:
            return 'List is empty'

        cur_index = 0
        cur = self.head.next
        prev = None

        while cur != None:
            if cur_index == index:
                if prev == None:
                    self.head.next = cur.next
                else:
                    prev.next = cur.next
                return
            prev = cur
            cur = cur.next
            cur_index += 1

        return 'Out of range'
    
    def sett(self,index,value):
        
        if self.head.next == None:
            return 'List is empty'
        
        cur = self.head.next
        cur_index = 0
        while cur != None:
            if cur_index == index:
                cur.data = value
                return True
            cur = cur.next
            cur_index += 1
        return 'Out of range'
    
    def __iter__(self):
        elem = []
        cur = self.head
        
        while cur.next != None:
            elem.append(cur.next.data)
            cur = cur.next
    
        return elem
    
    def printList(self):
        string = ''
        cur = self.head
        
        while cur.next != None:
            string += str(cur.next.data)
            string += ' '
            cur = cur.next
    
        return string
    
        

link = LinkedList()

link.addLast(5)
link.addLast(2)
link.addFirst(6)
link.addFirst(3)

print('-----------------')
print('First = ',link.getFirst())

print('-----------------')
print('Last = ',link.getLast())

print('-----------------')
print('Length = ',link.getSize())


print(link.__iter__())

print('-----------------')
print('First Node Removed')
link.removeFirst()

print(link.__iter__())
print('Length = ',link.getSize())

print('-----------------')
print('Last Node Removed')
link.removeLast()

print(link.__iter__())
print('Length = ',link.getSize())

print('-----------------')
print('Add ')
link.add(10)

print(link.__iter__())
print('Length = ',link.getSize())

print('-----------------')
print('Contains ')
print(link.contains(10))
print(link.__iter__())

print('-----------------')
print('Get ')
print(link.get(1))

print('-----------------')
print('Insert ')
link.insert(10,1)
print(link.__iter__())

print('-----------------')
print('Empty check ')
print(link.isEmpty())

print('-----------------')
print('Index of first value ')
print(link.indexOf(5))

print('-----------------')
print('Index of Last value ')
print(link.lastIndexOf(10))

print('-----------------')
print('Removing value ')
link.remove(10)
print(link.__iter__())

print('-----------------')
print('Remove at Index of')
link.removeAt(1)
print(link.__iter__())

print('-----------------')
print('Set value')
link.sett(1,9)
print(link.__iter__())

print('-----------------')
print(link.printList())


print('-----------------')
print('Clear ')
link.clear()

print(link.__iter__())
print('Length = ',link.getSize())
```
