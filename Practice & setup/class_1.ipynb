```python
class Employee:
    numberOfEmployee = 0
    raise_salary = 1.04
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '.@cpompany.com'
        Employee.numberOfEmployee += 1
        
    def fullName(self):
        return f"{self.first} {self.last}"
    
    def incres_sal(self):
        self.pay = int(self.pay * self.raise_salary)
        return self.pay 
    
    def __repr__(self):
        return f" Employee('{self.first}','{self.last}','{self.pay}')"
    
    def __str__(self):
        return f" {self.fullName()} - {self.email} "
    
    @classmethod
    def from_string(cls, emp_str):
        first , last , pay = emp_str.split('-')
        return cls(first , last , pay)
    
    @staticmethod 
    def check_weekend(day):
        if day.weekday() == 5 or day.weekday() == 6:
            return False
        return True

class Developer(Employee):
    raise_salary = 1.10
    
    def __init__(self, first, last, pay, prog_lang):
        super().__init__(first, last, pay)
        self.prog_lang = prog_lang
        
class Manager(Employee):
    
    def __init__(self, first, last, pay, employee = None ):
        super().__init__(first, last, pay)
        if employee == None:
            self.employee = []
        else:
            self.employee = employee
            
    def add_emp(self,emp):
        if emp not in self.employee:
            self.employee.append(emp)
            
    def remove_emp(self,emp):
        if emp in self.employee:
            self.employee.remove(emp)
            
    def print_emp(self):
        for x in self.employee:
            print('------->' + x.fullName())
        
        
        
        
emp_1 = Employee('Picchi', 'Pugni', 50000)
emp_2 = Employee('Bro', 'Pugni', 50000)

# print(emp_1)
print(emp_1.fullName())
Employee.fullName(emp_2)
# print(emp_2)

# 2nd concept class variable

emp_1.raise_salary = 1.10


print('Object Emp 1 = ',emp_1.incres_sal())
print('Object Emp 1 = ',emp_2.incres_sal())
print(Employee.numberOfEmployee)
# print('Object Emp 2 = ',emp_2.incres_sal)
# print('Class Employee = ',Employee.raise_salary)
# print('Class = ',Employee.__dict__)

#alter native Construtor 33333333333

#                              Note
# class has 3 methods regular methor which take instance , class method which take 
# class and staticmethod which is takes only argument
#                              Note

print('')
print('')
print('')
print('')
print('')
emp_3 = Employee.from_string('jamy-rahman-100000')
print(emp_3.__dict__)

import datetime
my_date =  datetime.date(2016,7,10)
print(Employee.check_weekend(my_date))





# Sub class 

print('')
print('')
print('')
print('')
print('               Part 4 start: Sub Class       ')

dev_1 = Developer('Yo', "sing", 50000, 'c')
dev_2 = Developer('Yo2', "sing2", 20000, 'js')

# print(help(Developer))
print(dev_1.pay)
dev_1.incres_sal()
print(dev_1.pay)
print(dev_1.prog_lang)

# mgr_1 = Manager('mangaer', 'last', 512313, ['jamy','jamy','jamy'])
# mgr_1.add_emp('salith')
# mgr_1.remove_emp('jamy')
# print(mgr_1.print_emp())
mgr_2 = Manager('mangaer', 'last', 512313, [dev_1,dev_2])
mgr_2.print_emp()
print(isinstance(mgr_2,Employee))
print(issubclass(Manager,Employee))



print('               Part 5 start: Special (Magic/Dunder) Methods       ')
print(emp_1)

print('               Part 6 start: Property Decorators - Getters, Setters, and Deleters       ')
print(emp_1)
```

```python
class NewEmployee :
    def __init__(self, first, last) :
        self.first = first
        self.last = last
    
    @property #getter
    def fullname (self):
        return f"{self.first} {self.last}"
    
    @fullname.setter
    def fullname(self, name):
        first , last = name.split(' ')
        self.first = first
        self.last = last
        
    @fullname.deleter
    def fullname(self):
        print('Delete name')
        self.first = None
        self.last = None
    
    @property
    def email(self):
        return f"{self.first}.{self.last}@gmail.com"

print('               Part 6 start: Property Decorators - Getters, Setters, and Deleters       ')
emp_1 = NewEmployee( 'John', 'Smith')
# emp_1.first = "Jim
emp_1.fullname = 'AL Jamy'
print (emp_1.first )
print( emp_1.email )
print (emp_1.fullname)
```
