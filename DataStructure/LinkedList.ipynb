# Linked List

```python
class Node:
    def __init__( self, data = None ):
        self.data = data
        self.next = None
        
        
class LinkedList:
    def __init__( self ):
        self.head = Node()
        
    def addFirst( self,data ):
        new_node = Node(data)
        new_node.next = self.head.next
        self.head.next = new_node
    
    def addLast( self,data ):
        self.add(data)
    
    def getFirst( self ):
        return self.head.next.data
    
    def getLast( self ):
        currentNode = self.head
        while currentNode.next != None:
            currentNode = currentNode.next
            
        return currentNode.data
    
    def removeFirst( self ):
        self.head = self.head.next
    
    def removeLast( self ):
        currentNode = self.head
        prevNode =''
        
        while currentNode.next != None:
            prevNode = currentNode
            currentNode = currentNode.next
        prevNode.next = None
    
    def add( self,data ):
        
        currentNode = self.head
        
        while currentNode.next != None:
            currentNode = currentNode.next
        
        currentNode.next = Node(data)
        
    
    def insert(self, index, data):
        
        currentNode = self.head
        prevNode = ''
        i = 0
        size = self.getSize()
        
        if index > size:
            print('Index is Out of Range')
            return 'Index is Out of Range'
        if index == size:
            self.add(data)
            return 
        
        while currentNode.next != None:
            prevNode = currentNode
            currentNode = currentNode.next
            
            if i == index:
                new_node = Node(data)
                prevNode.next = new_node
                new_node.next = currentNode
            i+=1
        
    def clear(self):
        self.head.next = None
    
    def contains(self,data):
        currentNode = self.head
        
        while currentNode.next != None:
            currentNode = currentNode.next
            if currentNode.data == data:
                return True
        return False
        
    def get(self,index):
        currentNode = self.head
        i = -1
        size = self.getSize()
        if index > size:
            print('Index is out of bound')
            return 'Index is out of bound'
        
        while currentNode.next != None:
            i +=1
            currentNode = currentNode.next
            
            if i==index:
                return currentNode.data
            
        
    
    def indexOf(self,data):
        currentNode = self.head
        i = -1
        while currentNode.next != None:
            currentNode = currentNode.next
            i+=1
            if currentNode.data == data:
                return i
            
        return -1
    
    def isEmpty(self):
        return False if self.head.next != None else True
            
    def lastIndexOf(self,data):
        
        if  self.contains(data) == False:
            return 'Item is not in the list'
        
        currentNode = self.head
        i = -1
        j = ''
        while currentNode.next != None:
            i += 1
            currentNode = currentNode.next
            
            if currentNode.data == data:
                j = i
        return j
    
    def getSize(self):
        currentNode =self.head
        i = 0
        while currentNode.next != None:
            i+=1
            currentNode = currentNode.next
        return i
    
    def remove(self,data):
        currentNode = self.head
        prevNode = ''
        if  self.contains(data) == False:
            return 'Item Is not In the Linked List'
        
        while currentNode.next != None:
            prevNode = currentNode
            currentNode = currentNode.next
            
            if currentNode.data == data:
                prevNode.next= currentNode.next
        
    
    def removeAt(self,index):

        size = self.getSize()
        
        if index> size:
            return 'Index is out of Range'
        
        currentNode = self.head
        prevNode = ''
        i = -1
        
        while currentNode.next != None:
            prevNode = currentNode
            currentNode = currentNode.next
            i+=1
            
            if i == index:
                prevNode.next = currentNode.next
    
    def sett(self,index,data):
        
        if index > self.getSize():
            return "Index is out of range"
        
        currentNode = self.head
        i = -1
        
        while currentNode.next != None:

            currentNode = currentNode.next
            i += 1
            if i == index:
                j= currentNode.data
                currentNode.data = data
                return f"{j} Has been replaced by {data}"
            
        
    
    def __iter__(self):
        linklist_array = []
        currentNode = self.head
        
        while currentNode.next != None:
            currentNode = currentNode.next
            linklist_array.append(currentNode.data)
            
        return linklist_array
    
homeMadeLinkedList = LinkedList()

print('-------------------  Add  -------------------')
homeMadeLinkedList.add(5)
homeMadeLinkedList.add(6)
print(homeMadeLinkedList.__iter__())

print('-------------------  Add First  -------------------')
homeMadeLinkedList.addFirst(10)
print(homeMadeLinkedList.__iter__())

print('-------------------  Add Last  -------------------')
homeMadeLinkedList.addLast(7)
print(homeMadeLinkedList.__iter__())

print('-------------------  Get First & Last  -------------------')
print( homeMadeLinkedList.getFirst() )
print( homeMadeLinkedList.getLast() )
print(homeMadeLinkedList.__iter__())

print('-------------------  Delete First  -------------------')

( homeMadeLinkedList.removeFirst() )
print(homeMadeLinkedList.__iter__())

print('-------------------  Delete Last  -------------------')
( homeMadeLinkedList.removeLast() )
print(homeMadeLinkedList.__iter__())

print('-------------------  Insert at  -------------------')
( homeMadeLinkedList.insert(2,20) )
( homeMadeLinkedList.insert(2,3) )
( homeMadeLinkedList.insert(0,10) )
print(homeMadeLinkedList.__iter__())

print('-------------------  Clear  -------------------')
homeMadeLinkedList.clear()
print(homeMadeLinkedList.__iter__())

print('-------------------  Adding items  -------------------')
homeMadeLinkedList.add(5)
homeMadeLinkedList.add(6)
homeMadeLinkedList.add(7)
homeMadeLinkedList.add(8)
print(homeMadeLinkedList.__iter__())

print('-------------------  Checking Contains  -------------------')
print(homeMadeLinkedList.contains(8))

print('-------------------  Getting Index Element  -------------------')
print(homeMadeLinkedList.get(3))

print('-------------------  Checking Element index  -------------------')
print(homeMadeLinkedList.indexOf(8))

print('-------------------  Checking List Empty or not  -------------------')
print(homeMadeLinkedList.isEmpty())

print('-------------------  Checking same element last index of Linked List -------------------')
homeMadeLinkedList.add(8)
print(homeMadeLinkedList.__iter__())
print(homeMadeLinkedList.lastIndexOf(8))

print('-------------------  Checking Size of Linked List  -------------------')
print(homeMadeLinkedList.getSize())

print('-------------------  Removing Item  -------------------')
(homeMadeLinkedList.remove(7))       
print(homeMadeLinkedList.__iter__())

print('-------------------  Removing Item at 1 index -------------------')
(homeMadeLinkedList.removeAt(1))       
print(homeMadeLinkedList.__iter__())

print('-------------------  Set item and returning item which is getting changed -------------------')
print((homeMadeLinkedList.sett(1,22)) )      
print(homeMadeLinkedList.__iter__())

print('-------------------  Iterating the Linked list -------------------')   
print(homeMadeLinkedList.__iter__())
```
